### matplotlib 基础

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)  # 前后全包 向量
x.shape
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)  # 折线图
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = "red",linestyle="--")
plt.xlim(-5,15)
plt.ylim(0,1.5)
```

```python
plt.plot(x,siny,label = "sin(x)")
plt.plot(x,cosy,color = "red",linestyle="--",label = "cos(x)")
plt.axis([-1,11,-2,2])
plt.xlabel("x axis")
plt.ylabel("y value")
plt.title("hello plt")
plt.legend()
```

### scatter plot  散点图

```python
plt.scatter(x,siny)
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color = "r")
```
