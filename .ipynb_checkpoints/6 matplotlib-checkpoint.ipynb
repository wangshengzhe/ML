### matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100) #向量 
```

```python
x
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color ='red',linestyle ='--')
plt.xlim(-5,15)
plt.ylim(0,1.5)
plt.show()
```

```python
plt.plot(x,siny,label = 'sin(x)')
plt.plot(x,cosy,color ='red',linestyle ='--',label = 'cos(x)')
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.title("hello matplotlib")
plt.show()
```

### Scatter Plot  散点图

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color = "red")
plt.show()
```

```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
plt.scatter(x,y,alpha = 0.4)
plt.show()
```
